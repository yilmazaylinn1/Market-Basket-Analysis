```python
#ihtiyaç duyulan kütüphaneler import edilir.
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
```

```python
#veri setinin ilk 5 değerleri
df.head() 
```

```python
# veri seti 532621 satırdan ve 8 sütundan oluşmaktadır.
df.shape
```

```python
#verinin columns isimleri
df.columns
```

```python
#Fatura numarası C ile başlayan ve fatura numarası olmayan verileri çıkarıyoruz.
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
df.head(10)
```

```python
#Veri setini Fransa’daki mağazalardan yapılan alışverişlerin fatura numaraları ve 
#bu alışverişlerde alınan ürünler ile ürün miktarını görebileceğimiz bir şekilde gruplandıralım.
#Veri setindeki boş verileri ‘0’ ile dolduralım. Gördüğümüz ‘0’ değerleri o ürünün alınmadığını belirtirken, 
#0’dan farklı olan değerler faturada bulunan ürün miktarını belirtir.

basket = (df[df['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
#gruplanan verinin ilk 10 sütunu
basket.head(10)
```

```python
#(Quantity) değişkeninin 0 olduğu durumlarda 0, 0’dan büyük olduğu durumlarda ise
#1 döndürecek basit bir fonksiyon

def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head(10)
```

```python
#Sıklıkla birlikte alınan ürünlerden destek parametresi %9’dan büyük olanları seçip 
#bu ürünlerden güven parametresi %80’den büyük olanları yeni bir değişkene atayalım.  
frequent_itemsets = apriori(basket_sets, min_support=0.09, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules.head()
```

```python
#Bir önceki basamakta minimum destek değeri %9 ve güven değeri %80 olan verileri incelemiştik. 
#Bu veriler içerisinde lift değeri 5’e eşit ve ya 5’ten büyük olanları seçelim. 
#Oluşturduğumuz bu tabloya bakarak birlikte satılma ihtimali oldukça yüksek olan ürünlerle ilgili yorum yapabiliriz.

rules[ (rules['lift'] >= 5) ]
```

```python
#aynı kuralı farklı ülkeler için de uyguayabiliriz. Burada Germany seçtim

basket2 = (df[df['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

basket_sets2 = basket2.applymap(encode_units)
basket_sets2.drop('POSTAGE', inplace=True, axis=1)
basket_sets2.head(10)
```

```python
frequent_itemsets = apriori(basket_sets2, min_support=0.09, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules.head()
```
